```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import math
import networkx.algorithms.community as nx_comm

#Set colour map for all plots
cmap = plt.get_cmap("tab10")
```

```python
#Epidemiological model
def run_SIR(Graph, beta, gamma, max_time):
    #Copy graph so no changes are made to original graph 
    G = Graph.copy()
    N = G.number_of_nodes()
    
    #Set infection status to susceptible for all nodes
    nx.set_node_attributes(G, 'S', 'status')
    
    #Randomly infect one person in the population
    patient_zero = np.random.randint(0,N-1)
    G.nodes[patient_zero]['status'] = 'I'
    
    #Initialise time at zero
    t = 0
    
    #Initialise arrays for storing number of nodes in each state
    T = np.arange(0,max_time)
    S = np.zeros(max_time)
    I = np.zeros(max_time)
    R = np.zeros(max_time)
    
    num_inf = 1
    while t < max_time:
        if num_inf > 0: 
            #Arrays for storing the indices of nodes that will have their state updated at the end of the timestep
            update_to_I = np.zeros(N)
            update_to_R = np.zeros(N)

            num_sus = 0 
            num_inf = 0
            for n in G.nodes:
                if G.nodes[n]['status'] == 'S':
                    num_sus +=1
                    #Count how many infected neighbours
                    i = 0
                    for neigh in G.neighbors(n):
                        if G.nodes[neigh]['status'] == 'I':
                            i += 1
                            
                    #If it has at least one inf neighbour, see if the node gets infected
                    if i > 0:
                        r = np.random.uniform()
                        inf_prob = 1 - math.exp(-beta*i)
                        if r < inf_prob:
                            update_to_I[n]=1
                elif G.nodes[n]['status'] == 'I':
                    num_inf += 1
                    #Recover with probability gamma
                    r = np.random.uniform()
                    if r < gamma:
                        update_to_R[n] = 1

            #Update all statuses
            for n,i in enumerate(update_to_I):
                if i == 1:
                    G.nodes[n]['status'] = 'I'
            for n,i in enumerate(update_to_R):
                if i == 1:
                    G.nodes[n]['status'] = 'R'

        
        S[t] = num_sus
        I[t] = num_inf
        R[t] = N - num_sus - num_inf
        t += 1
    
    return T, S, I, R, G
```

```python
p_list = np.arange(0, 1.01, 0.01)
#p_list = np.arange(0.15,0.3,0.001)
N = 100;
k = 10;

max_time = 500
num_sims = 100

av_clustWS = np.zeros(len(p_list)) #Average clustering coefficient
tot_infWS = np.zeros((len(p_list), num_sims)) #Total number of infections
I_arrayWS = np.zeros((len(p_list), max_time , num_sims))

R0 = 1.8
gamma=0.2
beta = R0*gamma/k

for i, p in enumerate(p_list):
    G_temp = nx.random_graphs.watts_strogatz_graph(N,k,p)
    av_clustWS[i] = nx.average_clustering(G_temp)
    for j in np.arange(0, num_sims):
        T, S, I, R, G_new = run_SIR(G_temp, beta, gamma, max_time)
        I_arrayWS[i, :, j] = I
        tot_infWS[i,j] = R[-1]
    
```

```python
av_tot_infWS = np.average(tot_infWS, 1)
```

```python
plt.scatter(p_list, av_clustWS)
plt.xlabel("$p$",fontsize=14)
plt.ylabel('Global clustering coefficient', fontsize=14)
plt.savefig("W-S_gcc.png")
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(p_list, av_clustWS,'o', color=cmap(0))
ax2.plot(p_list, av_tot_infWS,'o', color='r')
#ax2.errorbar(p_list, av_tot_inf, yerr=sd_tot_inf, fmt="o",color=cmap(1))
ax1.set_ylabel("Global clustering coefficient",color=cmap(0),fontsize=14)
ax2.set_ylabel("Outbreak size",color='r',fontsize=14)
ax1.set_xlabel("$p$",fontsize=14)
fig.savefig("WSp_gcc_aos.png")
```

```python
s = 25
sizes = [s, s, s, s]
num_comms = len(sizes)
N = sum(sizes)

maxd = k/(s-1)
d_list = np.arange(0,maxd, 0.004)
#d_list = np.arange(0.35,maxd,0.0004)


#Initialise arrays for storing data
av_degSBM = np.zeros(len(d_list)) #Average degree
av_clustSBM = np.zeros(len(d_list)) #Average clustering coefficient
modSBM = np.zeros(len(d_list))

tot_infSBM = np.zeros((len(d_list), num_sims)) #Total number of infections
I_arraySBM = np.zeros((len(d_list), max_time , num_sims))

comms = [set(range(0,s)),set(range(s,2*s)),set(range(2*s,3*s)),set(range(3*s,4*s))]

for i,d in enumerate(d_list):
    #Calculate non-diagonal values to keep average degree roughly constant
    od = (k - d*(sizes[0]-1))/((num_comms-1)*sizes[0])
    
    #Generate graph
    probs = np.ones((num_comms, num_comms))*od
    np.fill_diagonal(probs, d)
    G = nx.stochastic_block_model(sizes, probs)
    
    #Calculate average degree to check
    degrees = [d for u,d in G.degree()]
    av_degSBM[i] = np.average(degrees)
    
    #Calculate average clustering coefficient
    av_clustSBM[i] = nx.average_clustering(G)
    
    modSBM[i] = nx_comm.modularity(G, comms)
    
    for j in np.arange(0, num_sims):
        T, S, I, R, G_new = run_SIR(G, beta, 0.2, max_time)
        I_arraySBM[i, :, j] = I
        tot_infSBM[i,j] = R[-1]
        
    
av_tot_infSBM = np.average(tot_infSBM, 1)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.scatter(d_list, av_clustSBM, color = cmap(0))
ax2.scatter(d_list, av_tot_infSBM, color='r')
#ax2.errorbar(p_list, av_tot_inf, yerr=sd_tot_inf, fmt="o",color=cmap(1))
ax1.set_ylabel("Global clustering coefficient",color=cmap(0),fontsize=14)
ax2.set_ylabel("Outbreak size",color='r',fontsize=14)
ax1.set_xlabel("Probability of edges within community, $p_w$",fontsize=14)
fig.savefig("ERSBMp_gcc_aos.png")
```
